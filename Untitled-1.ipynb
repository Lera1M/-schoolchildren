### Визуализация данных

Для работы был взят набор данных с Kaggle, в котором представлены [результаты](https://www.kaggle.com/datasets/spscientist/students-performance-in-exams) экзаменов школьников США по нескольким предметам и факторы, которые могли на это повлиять.

Выведем несколько первых строк таблицы, чтобы взглянуть на данные:

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('work\students_performance.csv')
print(df.head())
sns.set_theme()
```

Выясняем результаты скольких человек хранятся в таблице:

```python
print(len(df))
```

#### Посмотрим на некоторые данные без графиков:

```python
pd.set_option('display.float_format', '{:.2f}'.format)
print(df.groupby(['race/ethnicity', 'gender'])['math score'].mean()) 
print(df.groupby(['race/ethnicity', 'gender'])['reading score'].mean()) 
print(df.groupby(['race/ethnicity', 'gender'])['writing score'].mean()) 
```

*По всем группам средний бал за экзамены по математике значительно выше у мальчиков, по остальным предметам - у девочек*

#### Рассмотрим некоторые зависимости, разделив учеников по половой принадлежности:

```python
sns.pairplot(df, hue="gender") 
plt.show()
```

*Действительно, графики на диагонали показывают, что мальчики показали себя лучше на экзамене по математике, а девочки в письме и чтении. Боковые графики это подтверждают.*

#### Посмотрим теперь на влияние некоторых факторов на результаты:

Используем прохождение подготовительных курсов

```python
sns.violinplot(x=df['test preparation course'], y=df['math score'], 
               hue=df['gender'], split=True)
plt.show()
sns.violinplot(x=df['test preparation course'], y=df['reading score'], 
               hue=df['gender'], split=True)
plt.show()
sns.violinplot(x=df['test preparation course'], y=df['writing score'], 
               hue=df['gender'], split=True)
plt.show()
```

*Подготовительные курсы по лбому предмету помогают ученикам вне зависимости от их пола*

Исследуем питание школьников:

```python
sns.violinplot(x=df['lunch'], y=df['reading score'], #чем лучше еда, тем лучше результат
               hue=df['gender'], split=True)
plt.show()
```

*Качественное питание значительно поднимает средний балл за экзамен (вне зависимости от пола)*

Рассмотрим этническую принадлежность (раз уж она есть в таблице):

```python
sns.barplot(data=df, x="race/ethnicity", y="math score") 
plt.show()
sns.barplot(data=df, x="race/ethnicity", y="reading score") 
plt.show()
sns.barplot(data=df, x="race/ethnicity", y="writing score") 
plt.show()
```

*Средний балл действительно зависит и от этической принадлежности. Группа А явно справилась хуже, а группа Е лучше остальных*

Посмотрим на образование родителей:

```python
sns.stripplot(data=df, y="parental level of education", x="math score") 
plt.show()
sns.stripplot(data=df, y="parental level of education", x="reading score")
plt.show()
sns.stripplot(data=df, y="parental level of education", x="writing score")
plt.show()
```

*Можно заметить, что у тех школьников, чьи родители имеют высшее образование, нет слишком плохих результатов*

Проверим корреляцию между результатами по математике и по письму

```python
df['math score'].corr(df['writing score'])
```

Коэффицент достаточно близок к единице, а значит есть корреляция. Построим график и отобразим результат наглядно

```python
sns.lineplot(data=df, y=df['math score'], x=df['writing score'], color='indigo')
```
